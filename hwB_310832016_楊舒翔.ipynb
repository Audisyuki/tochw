```python
import matplotlib.pyplot as plt
import numpy as np

from scipy.ndimage import gaussian_filter
def blur(img, dpi):
    return gaussian_filter(img, (3,3,0)), -0, 0

def plotx(*a, linewidth=1, **k):
    plt.plot(*a, **k, linewidth=15*linewidth, alpha=0.3, agg_filter=blur)
    plt.plot(*a, **k, linewidth=7*linewidth, alpha=0.8, agg_filter=blur)
    plt.plot(*a, **k, linewidth=linewidth)
```
